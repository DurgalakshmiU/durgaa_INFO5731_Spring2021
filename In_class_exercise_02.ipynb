(1) Write a Python program to find the duplicate elements in a given array of integers. Return -1 If there are no such elements. (4 points)

```python
def find_duplicateelements(arr):
        numbers_set = set ()
        no_duplicateelements=-1

        for i in range(len(arr)):
        
          if arr[i] in numbers_set:
            return arr[i]

          else:
           numbers_set.add(arr[i]) 


       
        return no_duplicateelements 

print(find_duplicateelements([3,6,7,9,1,1,10,2,5]))   
print(find_duplicateelements([1,5,3,7,8,4,2,9,6]))      


      


```

(2) Write a Python program to select all the Sundays of a specified year. (4 points)



```python
from datetime import date,timedelta
def all_sundays(year):
  dt=date(year,1,1)
  dt+=timedelta(days= 6 - dt.weekday())
  while dt.year == year:
     yield dt
     dt += timedelta(days = 7)

for s  in all_sundays(2022):
     print(s)


```
